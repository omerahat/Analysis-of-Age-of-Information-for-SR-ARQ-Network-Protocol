```python
file_path = "AoI.csv"
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv(file_path)

# Plotting the line graph
plt.figure(figsize=(200, 10))
plt.plot(data.index, data['AoI'], marker='o', color='blue')
plt.title('Line Graph of AoI Data')
plt.xlabel('Index')
plt.ylabel('AoI')
plt.grid(True)
plt.savefig('AoI_line_graph.png')
plt.show()
```
